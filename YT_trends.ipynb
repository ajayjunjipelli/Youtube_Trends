```python
df['category'] = df['category'].fillna(df['category'].mode()[0])
```

```python
df['channel_type']=df['channel_type'].fillna(df['channel_type'].mode()[0])
```

```python
df['Country'] = df['Country'].fillna('Unknown')
df['Abbreviation'] = df['Abbreviation'].fillna('Unknown')
```

```python
df['created_month']=df['created_month'].fillna('Unkhown')
```

```python
df.isnull().sum()
```

```python
for col in Numerical1:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isnull().sum()
```

```python
df['Urban_population'] = df['Urban_population'].fillna(df['Urban_population'].median())
```

```python
df['Latitude'] = df['Latitude'].fillna(df['Latitude'].median())
```

```python
df['Longitude'] = df['Longitude'].fillna(df['Longitude'].median())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Title Sentiment Distribution
plt.figure(figsize=(8,4))
sns.histplot(df['Title'], bins=30, kde=True)
plt.title('Title Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Count')
plt.show()
```

```python
avg_views = df.groupby('category')['video views'].mean().sort_values(ascending=False)
plt.figure(figsize=(10,5))
avg_views.plot(kind='bar', color='skyblue')
plt.title('Average Views by Category')
plt.ylabel('Average Views')
plt.xticks(rotation=45)
plt.show()
```

#  Distribution of Subscribers 

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['subscribers'], bins=50, kde=True)
plt.title('Distribution of Subscribers')
plt.xlabel('Subscribers')
plt.ylabel('Frequency')
plt.show()
```

# Top 10 Channels by Video Views

```python
top_views = df.sort_values('video views', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='video views', y='Youtuber', data=top_views, palette='viridis')
plt.title('Top 10 Youtubers by Video Views')
plt.xlabel('Video Views')
plt.ylabel('Youtuber')
plt.show()
```

# Average Subscribers by Category

```python
avg_subs = df.groupby('category')['subscribers'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x=avg_subs.values, y=avg_subs.index, palette='magma')
plt.title('Average Subscribers by Category')
plt.xlabel('Average Subscribers')
plt.ylabel('Category')
plt.show()
```

#  Sentiment Distribution of Titles 

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Title'], bins=30, kde=True, color='teal')
plt.title('Title Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```

#  Video Views vs Subscribers (Scatter Plot)

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='subscribers', y='video views', hue='category', data=df, alpha=0.7)
plt.title('Video Views vs Subscribers by Category')
plt.xlabel('Subscribers')
plt.ylabel('Video Views')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

# Correlation Heatmap Between Key Metrics

```python
metrics = df[['subscribers', 'video views', 'video_views_for_the_last_30_days',
              'lowest_yearly_earnings', 'highest_yearly_earnings',
              'subscribers_for_last_30_days']]

plt.figure(figsize=(10, 6))
sns.heatmap(metrics.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Between Channel Metrics')
plt.show()
```

# Region-wise (Country-wise) Comparison of Views

```python
top_countries = df['Country'].value_counts().head(5).index
subset = df[df['Country'].isin(top_countries)]

plt.figure(figsize=(12, 6))
sns.boxplot(x='Country', y='video views', data=subset)
plt.title('Video Views by Country (Top 5)')
plt.xlabel('Country')
plt.ylabel('Video Views')
plt.yscale('log')  # if skewed
plt.show()
```

# Channel Creation Trends Over the Years

```python
year_counts = df['created_year'].value_counts().sort_index()

plt.figure(figsize=(10, 5))
sns.lineplot(x=year_counts.index, y=year_counts.values, marker='o')
plt.title('Channels Created Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Channels')
plt.grid(True)
plt.show()
```
